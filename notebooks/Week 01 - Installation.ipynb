## MAC
### Python
Python comes pre-installed on Mac OS X, but due to Apple's release cycle, it's often one or even two years old. The overwhelming recommendation of the "MacPython" community is to upgrade your Python by downloading and installing a newer version from the Python standard release page.

<ol>
    <li>Download the latest Python2 release for your system to your desktop:</li>
    <ul>
        <li> [<a href="http://www.python.org/ftp/python/2.7.5/python-2.7.5-macosx10.3.dmg" target="_blank">Python2 32-bit</a> (MacOS X 10.3+)] </li>
        <li> [<a href="http://www.python.org/ftp/python/2.7.5/python-2.7.5-macosx10.6.dmg" target="_blank">Python2 64-bit</a> (MacOS X 10.6+)] </li>
    </ul>
    <li>Open a terminal: <i>/Applications/Utilities/Terminal</i></li>
    <li>Navigate to Desktop by typing <cb>cd ~/Desktop/</cb></li>
    <li>Install Python2 by running either</li>
    <cb>hdiutil mount python-2.7.5-macosx10.3.dmg</cb> for 32-bit, or<br />
    <cb>hdiutil mount python-2.7.5-macosx10.6.dmg</cb> for 64-bit.<br/>
    and then, <br />
    <cb>sudo installer -pkg "/Volumes/Python 2.7.5/Python.mpkg" -target /</cb>
</ol>
### IPython and extras
The program <i>easy_install</i> comes with python and can be used to install additional Python2 packages:

<cb>sudo easy_install ipython</cb><br />
<cb>sudo easy_install numpy</cb><br />
<cb>sudo easy_install scipy</cb><br />
<cb>sudo easy_install pandas</cb><br />
<cb>sudo easy_install tornado</cb><br />
<cb>sudo easy_install pyzmq</cb><br />
