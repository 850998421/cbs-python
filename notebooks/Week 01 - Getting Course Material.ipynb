### Linux

All you need is git. If you are running ubuntu or debian you get it by typing 

<cb>sudo apt-get install git</cb>

if you are running fedora you can type

<cb>sudo yum install git</cb>

Now set up your git account on your system

<cb>git config --global user.name "<span style="color: red;">git-username</span>"</cb>
<cb>git config --global user.email "<span style="color: red;">your@student.mail.dk</span>"</cb>
<cb>git config --global credential.helper 'cache --timeout=3600'</cb>

Now you need to choose a directory to store the course materials, eg <cb>/home/christian/Documents/courses/</cb>. Navigate to the folder:

<cb>cd /home/christian/Documents/courses</cb>

now clone the cbs-python github 

<cb>git clone https://github.com/skjoedt/cbs-python.git</cb>

If things worked out as expected you should have a folder called <cb>cbs-python</cb> from where you called the <cb>git clone</cb> command.
